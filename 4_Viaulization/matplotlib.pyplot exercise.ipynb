```python
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## 1. single plot

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
```

```python
plt.figure(1, figsize=(10,5))
sns.countplot(x='a', data=data, color='g')

plt.title('main title test')
plt.xlabel('a')
plt.ylabel('cnt')
plt.grid(True)


plt.show()
```

## 2. subplots

```python
fig=plt.figure(2, figsize=(15,6))
((ax1,ax2,ax3)),((ax4,ax5,ax6))=fig.subplots(nrows=2, ncols=3)

ax1=sns.boxplot(x='c', data=data)
ax1.set(xlabel='a', title="test")
ax1.set_xtics([10,50])
```
