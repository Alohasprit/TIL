```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

### 1. 이산형 pairplot

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris);
```

### 2. 연속형 multiple pair plot)

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6);
```
