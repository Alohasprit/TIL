```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings

warnings.filterwarnings(action='ignore')

%matplotlib inline
```

```python
df=pd.DataFrame(dict(time=np.arange(500), value=np.arange(500).cumsum()))
```

```python
g=sns.relplot(x='time', y='value', kind='line', data=df)
g.fig.autofmt_xdate()
```

#### 동일한 x 값에 대한 y 값은 평균 그리고 평균을 중심으로 95% 신뢰 구간 데이터를 아래와 같이 음영처리로 표시한다

```python
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri)
```

```python
# 신뢰구간을 없앨 수 있는 인자 'ci=None' 도 있음
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

```python
# 신뢰구간 대신 표준편차를 표현할 수도 있는 인자도 있다 'kind=line'
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

```python
# estimator를 끄고 실제 중복된 값을 모두 표현할 수도 있음 'estimator=None'
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri);
```

```python
# lineplot 또한 hue 인자를 사용할 수 있다
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

```python
#style 인자를 사용할 수 있음
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            kind="line", data=fmri);
```

```python
# row와 column 인자에 카테고리 변수를 대입하여 그래프를 다양하게 그릴 수 있음
sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri)
```

```python
# variable이 많은 컬럼을 비교할때는 아래와 같이 컬럼을 wrap해서 비교하는 방법도 있다
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'frontal'"));
```
