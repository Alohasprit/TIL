```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

### 1. 정규분포(Normal Distribution)

정규 분포는 가우시안(Gaussian distribution) 분포라고도 불리며 평균 u를 중심으로 관측값 x에 대한 확률분포p(x)의 bell 모양 곡선을 보이며, 평균 u를 중심으로 대칭이면 정규분포, 비대칭은 비정규분포로 정의된다.

표기 방법는 N(u,a**2)

### 1-1 정규분포 데이터 생성

```python
np.random.seed(42)

data = np.random.randn(5)
data
```

```python
f, ((ax1),(ax2)) = plt.subplots(nrows=1,ncols=2)
f.set_size_inches(10,5)

sns.scatterplot(x=data, y=data, ax=ax1)
sns.distplot(data, ax=ax2)
```

```python
data.mean(), data.std()
```

### 1-2 정규분포 데이터의 표준화(Standarzation)
정규분포의 성격을 유지하면서 표준화N(0,1) 됨

```python
data = (data-data.mean())/data.std()
data
```

```python
f, ((ax1),(ax2)) = plt.subplots(nrows=1,ncols=2)
f.set_size_inches(10,5)

sns.scatterplot(x=data, y=data, ax=ax1)
sns.distplot(data, ax=ax2)
```

```python
data.mean(), data.std()
```

### 1-2 정규분포 데이터의 정규화(Normalization)

```python
np.random.seed(43)
data = np.random.randn(5)
data
```

```python
data = (data - data.min())/(data.max()-data.min())
data
```

```python
f, ((ax1),(ax2)) = plt.subplots(nrows=1,ncols=2)
f.set_size_inches(10,5)

sns.scatterplot(x=data, y=data, ax=ax1)
sns.distplot(data, ax=ax2)
```

```python
data.mean(), data.std()
```

### 2. 비정규 분포

#### 비정규분포
 비정규분포

### 2-1 비정규 분포 데이터 생성

```python
data = np.array([1,1,2,2,2,3,4,5])
data
```

```python
m = data.mean()
m
```

```python
a = data.std()
a
```

```python
f, ((ax1),(ax2)) = plt.subplots(nrows=1,ncols=2)
f.set_size_inches(10,5)

sns.scatterplot(x=data, y=data, ax=ax1)
sns.distplot(data, ax=ax2)
```

### 2-2 비정규 분포 데이터의 표준화(Standarzation)
비정규 분포 데이터를 표준화 하더라도 평균을 중심으로 비대칭성(비정규분포)는 유지되며, 다만 평균 u = 0 , 표준편차 a = 1로 변경된다.

```python
data2 = []

for i in data:
    e = (i-m) / a
    data2.append(e)
    print (e)

data2 = np.array(data2)
```

```python
f, ((ax1),(ax2)) = plt.subplots(nrows=1,ncols=2)
f.set_size_inches(10,5)

sns.scatterplot(x=data2, y=data2, ax=ax1)
sns.distplot(data2, ax=ax2)
```

```python
data2.mean()
```

```python
data2.std()
```

### 2-2-1 비정규 분포의 표준화 후, 정규화를 한다면?
비정규성은 그대로 유지된 채, 관측값만 0<x<1 범위로 re-scaring 됨

```python
data2
```

```python
data2_1 = (data2 - data2.min()) / (data2.max()-data2.min())
data2_1
```

```python
f, ((ax1),(ax2)) = plt.subplots(nrows=1,ncols=2)
f.set_size_inches(10,5)

sns.scatterplot(x=data2_1, y=data2_1, ax=ax1)
sns.distplot(data2_1, ax=ax2)
```

```python
data2_1.mean(), data2_1.std()
```

### 2.3 비정규 분포의 정규화(normalization)

관측 데이터 X의 분포를 0<x<1 로 Scaring 해 줌. 분포의 비대칭성은 유지
이미지 데이터의 각 픽셀 값(0 to 255)의 정규화에 주요 쓰이며,
batch normalization(activation func 이전에 입력 데이터를 0 to 1 값으로 normalizing) 하는데 쓰임 --> gradient vanishing 방지

```python
data
```

```python
data.max(), data.min()
```

```python
data3 = (data - data.min())/(data.max()-data.min())
data3
```

```python
f, ((ax1),(ax2)) = plt.subplots(nrows=1,ncols=2)
f.set_size_inches(10,5)

sns.scatterplot(x=data3, y=data3, ax=ax1)
sns.distplot(data3, ax=ax2)
```

```python
data3.mean(), data3.std()
```

### 2-3-1 비정규 분포 데이터를 정규화 그리고 표준화를 진행한다면?

비정규분포의 비정규성(평균을 중심으로 비대칭 확률분포)는 변화지 않음

```python
data4 = (data3 - data3.mean())/data3.std()
data4
```

```python
f, ((ax1),(ax2)) = plt.subplots(nrows=1,ncols=2)
f.set_size_inches(10,5)

sns.scatterplot(x=data4, y=data4, ax=ax1)
sns.distplot(data4, ax=ax2)
```

```python
data4.mean(), data4.std()
```
