```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.optim as optim
import  torch.nn.init as init
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = init.uniform(torch.FloatTensor(1000,1), -10,10)
y = init.uniform(torch.FloatTensor(1000,1), -20,20)
z = x**2 + y**2
```

```python
input_data=Variable(torch.cat([x,y], dim=1))

model = nn.Sequential(
                nn.Linear(2,20),
                nn.ReLU(),
                nn.Linear(20,20),
                nn.ReLU(),
                nn.Linear(20,5),
                nn.ReLU(),
                nn.Linear(5,5),
                nn.ReLU(),
                nn.Linear(5,1),
)
```

```python
criterion = nn.L1Loss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
for i in range(5000):
    optimizer.zero_grad()
    y_pred = model(input_data)
    loss = criterion(y_pred, Variable(z))
    loss.backward()
    optimizer.step()
    
    if i % 50 == 0:
        print(i, loss)
        
    
    
```
